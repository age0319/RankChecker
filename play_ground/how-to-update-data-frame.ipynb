```python
import pandas as pd
```

```python
# List of Tuples
students = [ ('jack', 34, 'Sydeny' , 'Australia') ,
('Riti', 30, 'Delhi' , 'India' ) ,
('Vikas', 31, 'Mumbai' , 'India' ) ,
('Neelu', 32, 'Bangalore' , 'India' ) ,
('John', 16, 'New York' , 'US') ,
('Mike', 17, 'las vegas' , 'US')  ]

#Create a DataFrame object
df = pd.DataFrame(students, columns = ['Name' , 'Age', 'City' , 'Country']) 
df
```

```python
#　インデックスとカラム名で書き換える
df.loc[0,"Age"] = 10
df
```

```python
#行と列のインデックスで書き換える
df.iloc[0,2] = "Gold Coast"
df
```

```python
# 名前を検索して年齢を書き換える
user_input = "Mike"
df.loc[df[df["Name"] == user_input].index, 'Age'] = 72
df
```

```python
# ２番目の行を別の行に入れ替える
students_new_comer = ['Obi', '32','Tokyo','Japan']
df.loc[2]=students_new_comer
df
```

```python
new_comer = [('ShinHua', '3','HongKong','China')]
new_df = pd.DataFrame(new_comer, columns = ['Name' , 'Age', 'City' , 'Country']) 
df = df.append(new_df)
df
```

```python
new_comer = [('ShinHua', 'hoge','HongKong','China')]
new_df = pd.DataFrame(new_comer, columns = ['Name' , 'Age', 'City' , 'Country']) 
df = df.append(new_df)
df
```
