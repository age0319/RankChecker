```python
import pandas as pd
# List of Tuples
students = [ ('jack', 34, 'Sydeny' , 'Australia') ,
('Riti', 30, 'Delhi' , 'India' ) ,
('Vikas', 31, 'Mumbai' , 'India' ) ,
('Neelu', 32, 'Bangalore' , 'India' ) ,
('John', 16, 'New York' , 'US') ,
('Mike', 17, 'las vegas' , 'US')  ]

#Create a DataFrame object
df = pd.DataFrame(students, columns = ['Name' , 'Age', 'City' , 'Country']) 
df
```

```python
# johnを探す
df[df["Name"] == "John"]
```

```python
# 名前にaが含まれるデータを探す
df[df.Name.str.contains("a")]
```

```python
# 名前にeeが含まれない人を探す
df[(~df.Name.str.contains("ee"))]
```

```python
# 年齢が30歳以上の人を探す
df[df["Age"] >= 30]
```

```python
# インドで31歳以上の人を探す
df[(df["Age"] >= 31) & (df["Country"] == "India")]
```

```python
# インドもしくはアメリカの人を探す
df[(df["Country"] == "India") | (df["Country"] == "US")]
```
