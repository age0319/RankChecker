```python
import pandas as pd

# List of Tuples
students = [ ('jack', 34, None,None, 'Sydeny' , 'Australia') ,
('Riti', 30, None,None,'Delhi' , 'India' ) ,
('Vikas', 5, None,None,'Mumbai' , 'India' ) ,
('Neelu', 32, None, None,'Bangalore' , 'India' ) ,
('John', 16, None, None,'New York' , 'US') ,
('Mike', "50+", None, None,'las vegas' , 'US')  ]

#Create a DataFrame object
df = pd.DataFrame(students, columns = ['Name' , 'Rank', 'Pre', 'Diff','City' , 'Country']) 

df['Rank is String?'] = [isinstance(x, int) for x in df.Rank]
df['Pre is String?'] = [isinstance(x, int) for x in df.Pre]

print(df)

def update(index,newrank):
    pre = df.loc[index, "Rank"]
    df.loc[index,"Pre"] = pre
    df.loc[index,"Rank"] = newrank

    if (isinstance(newrank, int) == True) & (isinstance(pre, int) == True):
            df.loc[index,"Diff"]  = pre - newrank
    else: 
            df.loc[index,"Diff"]  = None
```

```python
# ランキングが上がった時のテスト
newrank = 1
index = 0

update(index, newrank)

df
```

```python
# ランキングが下がった時のテスト
newrank = 31
index = 4

update(index, newrank)

df
```

```python
# ランキング外➡︎ランキング内
newrank = 31
index = 5

update(index, newrank)

df
```

```python
# ランキング内➡︎ランキング外
newrank = "50+"
index = 2

update(index, newrank)

df
```
